# **4. Data Cleaning**

**4.1. Cleaning CarName Feature**

```python
df.head()
```

We need only car company name not the car model name so we remove the car model name

```python
Company_name=df["CarName"].apply(lambda x:x.split(" ")[0])
df.insert(2,"Car_company",Company_name)
df.drop(columns=['CarName'],inplace=True)
```

```python
df.head()
```

```python
df["Car_company"].unique()
```

we found out that there are some spelling errors
1. maxda=mazda
2. Nissan=nissan
3. porsche=porcshce
4. toyota=toyouta
5. vokswagen=volkswagen=vw

so we make a function to correct the company names


```python
def replace(a,b):
    df["Car_company"].replace(a,b,inplace=True)

replace('maxda','mazda')
replace('porcshce','porsche')
replace('toyouta','toyota')
replace('vokswagen','volkswagen')
replace('vw','volkswagen')
```

# **6.Data Wrangling**

```python
z = round(df.groupby(["Car_company"])["price"].agg(["mean"]),2).T
z
```

```python
df = df.merge(z.T,how="left",on="Car_company")
```

```python
bins = [0,10000,20000,40000]
cars_bin=['Budget','Medium','Highend']
df['CarsRange'] = pd.cut(df['mean'],bins,right=False,labels=cars_bin)
df.head()
```

# **7. Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
new_df = df[['fueltype','aspiration','doornumber','carbody','drivewheel','enginetype','cylindernumber','fuelsystem'
             ,'wheelbase','carlength','carwidth','curbweight','enginesize','boreratio','horsepower','citympg','highwaympg',
             'price','CarsRange']]
new_df.head()
```

**Creating dummies variable for categorical Features**

```python
new_df = pd.get_dummies(columns=["fueltype","aspiration","doornumber","carbody","drivewheel","enginetype",
                                "cylindernumber","fuelsystem","CarsRange"],data=new_df)
new_df.head()
```

```python
num_cols = ['wheelbase','carlength','carwidth','curbweight','enginesize','boreratio','horsepower',
            'citympg','highwaympg']

new_df[num_cols] = scaler.fit_transform(new_df[num_cols])
new_df.head()
```

# **8.Feature Selection**

**Splitting Data**

```python
x= new_df.drop(columns=["price"])
y = new_df["price"]
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=60)
```

```python
print("x_train - >  ",x_train.shape)
print("x_test - >  ",x_test.shape)
print("y_train - >  ",y_train.shape)
print("y_test - >  ",y_test.shape)
```

# **9.Model Implementation And Evaluation**

**9.1.Decision Tree Regression**

9.1.1. Model Implementaion

```python
from sklearn.tree import DecisionTreeRegressor as DTR
reg1= DTR(max_depth=2)
reg2=DTR(max_depth=5)
reg1.fit(x_train,y_train)
reg2.fit(x_train,y_train)
```

```python
##Accuracy
print(reg1.score(x_test,y_test))
reg2.score(x_test,y_test)
```

```python
#testing the model
y_predicted_DTR = reg2.predict(x_test)
y_predicted_DTR
```

9.1.2. Model Evaluation

Mean Squared error(MSE)

```python
from sklearn.metrics import mean_squared_error as mse
np.sqrt(mse(y_test,y_predicted_DTR))
```

```python
from sklearn.model_selection import cross_val_score as cvs
cvs(reg2,x_train,y_train,cv=10)
```

**9.2. SVM Regression**

9.2.1 Model Implementaion

```python
from sklearn.svm import SVR
svr_reg=SVR(kernel='rbf')
svr_reg.fit(x_train,y_train)
```

```python
##Accuracy
svr_reg.score(x_test,y_test)
```

```python
y_pred_svr = svr_reg.predict(x_test)
y_pred_svr
```

9.2.2. Model Evaluation

Mean Squared error(MSE)

```python
np.sqrt(mse(y_test,y_pred_svr))
```

Cross Validation score

```python
cvs(svr_reg,x_train,y_train,cv=10)
```

We used two Regression models Decision tree Regression and SVM Regression
we found that the Decision tree regression has the highest accuracy and the model performs well
